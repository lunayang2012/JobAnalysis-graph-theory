```python
from pathlib import Path
import sys
import subprocess
import os

print("Python executable:", sys.executable)
```

```python
# ========= MODE =========
MODE = "precise"   # or "fast"

# ========= PATHS =========
PROJECT_ROOT = Path(r"C:\Users\astef\anaconda3\AI and Analytics Research Group\Job_Skills")

BEFORE_PATH = PROJECT_ROOT / "job_skills" / "data" / "inputs" / "Main_Data_File.xlsx"
AFTER_PATH  = PROJECT_ROOT / "job_skills" / "data" / "processed" / "CleanJobs_Final_v2.0.xlsx"          

# ========= WHAT TO RUN =========
RUN_FULL_PIPELINE      = False
RUN_PREPROCESS_ONLY    = False
RUN_DQ_REPORT          = True
RUN_BERTOPIC_ONLY      = False
RUN_LABEL_TOPICS_LLM   = False
RUN_EXTRACT_SKILLS     = False

# ========= EXTRA FLAGS (optional) =========
EXTRA_ARGS_RUN_ALL        = []
EXTRA_ARGS_PREPROCESS     = []
EXTRA_ARGS_DQ_REPORT      = []
EXTRA_ARGS_BERTOPIC       = []
EXTRA_ARGS_LLM_LABEL      = []
EXTRA_ARGS_SKILL_EXTRACT  = []
```

```python
def run_cmd(cmd):
    print("\n>>> RUNNING:")
    print(" ".join(cmd))
    print("-" * 80)

    p = subprocess.Popen(
        cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    out, err = p.communicate()
    if out:
        print("\n--- STDOUT ---\n", out)
    if err:
        print("\n--- STDERR ---\n", err)

    if p.returncode != 0:
        raise RuntimeError(f"Command failed with exit code {p.returncode}")
```

```python
if RUN_FULL_PIPELINE:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "run-all",
        "--mode", MODE,
        "--input", str(BEFORE_PATH),
        *EXTRA_ARGS_RUN_ALL,
    ]
    run_cmd(cmd)
```

```python
if RUN_PREPROCESS_ONLY:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "preprocess",
        "--mode", MODE,
        "--input", str(BEFORE_PATH),
        *EXTRA_ARGS_PREPROCESS,
    ]
    run_cmd(cmd)
```

```python
if RUN_DQ_REPORT:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "dq-report",
        "--mode", MODE,
        "--before", str(BEFORE_PATH),
        "--after", str(AFTER_PATH),
        *EXTRA_ARGS_DQ_REPORT,
    ]
    run_cmd(cmd)
```

```python
if RUN_BERTOPIC_ONLY:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "bertopic",
        "--mode", MODE,
        *EXTRA_ARGS_BERTOPIC,
    ]
    run_cmd(cmd)
```

```python
if RUN_LABEL_TOPICS_LLM:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "label-topics-llm",
        "--mode", MODE,
        *EXTRA_ARGS_LLM_LABEL,
    ]
    run_cmd(cmd)
```

```python
if RUN_EXTRACT_SKILLS:
    cmd = [
        sys.executable, "-m", "job_skills.pipeline.cli",
        "extract-skills",
        "--mode", MODE,
        *EXTRA_ARGS_SKILL_EXTRACT,
    ]
    run_cmd(cmd)
```
